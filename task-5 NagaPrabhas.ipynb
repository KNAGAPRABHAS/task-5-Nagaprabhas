```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load dataset
df = pd.read_csv('train_and_test2.csv')

# Step 3: Basic info and preview
print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nFirst 5 Rows:\n", df.head())

# Step 4: Descriptive statistics
print("\nDescriptive Statistics:\n", df.describe(include='all'))

# Step 5: Value counts for categorical and low-unique columns
print("\n=== Value Counts for Object-Type Columns ===")
obj_cols = df.select_dtypes(include='object').columns
if len(obj_cols) > 0:
    for col in obj_cols:
        print(f"\nValue counts for '{col}':\n", df[col].value_counts())
else:
    print("No object-type (categorical) columns found.")

print("\n=== Value Counts for Columns with < 10 Unique Values ===")
for col in df.columns:
    if df[col].nunique() < 10:
        print(f"\nValue counts for '{col}':\n", df[col].value_counts())


```

```python
# For object-type (categorical) columns
print("\n=== Value Counts for Object-Type Columns ===")
obj_cols = df.select_dtypes(include='object').columns
if len(obj_cols) > 0:
    for col in obj_cols:
        print(f"\nValue counts for '{col}':\n", df[col].value_counts())
else:
    print("No object-type (categorical) columns found.")

# For numeric columns with fewer than 10 unique values
print("\n=== Value Counts for Columns with < 10 Unique Values ===")
for col in df.columns:
    if df[col].nunique() < 10:
        print(f"\nValue counts for '{col}':\n", df[col].value_counts())
```

```python
sns.pairplot(df)
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#histogram

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols].hist(figsize=(15, 10), bins=30, edgecolor='black')
plt.suptitle("Histograms of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
#boxplots

plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols):
    plt.subplot((len(num_cols) + 2) // 3, 3, i + 1)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
#scatterplot

target_col = None
for col in df.columns:
    if col.lower() in ['target', 'label', 'outcome']:
        target_col = col
        break

if target_col:
    print(f"\nUsing '{target_col}' as target for scatterplots.")
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(num_cols):
        if col != target_col:
            plt.subplot((len(num_cols) + 2) // 3, 3, i + 1)
            sns.scatterplot(x=df[col], y=df[target_col])
            plt.title(f"{col} vs {target_col}")
    plt.tight_layout()
    plt.show()
else:
    print("\nNo obvious target column found for scatterplots.")
```
